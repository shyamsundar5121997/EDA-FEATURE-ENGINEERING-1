# Change the column name {0} to rating count.

```python
rating=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
rating.head()
```

#Observation:
1.when Rating is netween 4.5 to 4.9 ---------> Excellent.
2.when Rating are between 4.0 to 3.4---------->Very good.
3.when Rating are between 3.5 to 3.9---------->Good.
4,when Rating are between 3.0 to 3.4---------->Average.
5.when Rating are between 2.5 to 2.9---------->Average.
6.when Rating are between 2.0 to 2.4----------->Poor.

```python
sns.barplot(x='Aggregate rating',y='Rating count',data=rating)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=rating)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=rating)
```

Observation:
1. Not Rated count is very high.
2.Maxmimum number of rating are between 2.5 to 3.4.
